# Atividades m√≠nimas: 65 pts

```python
import numpy as np
import matplotlib.pyplot as plt
#Fun√ß√£o para plotar curvas gaussianas com diferentes centros e larguras
def plot_gaussians():
    x = np.linspace(-10, 10, 400)
    centers = [-3, 0, 3]
    widths = [0.5, 1, 2]

    plt.figure(figsize=(8, 5))
    for c, w in zip(centers, widths):
        y = np.exp(-((x - c) ** 2) / (2 * w ** 2))  # Equa√ß√£o da fun√ß√£o gaussiana
        plt.plot(x, y, label=f"center={c}, width={w}")

    plt.title("Fun√ß√£o Gaussiana")
    plt.legend()
    plt.show()

def plot_sinc():
    x = np.linspace(-10, 10, 400)
    y = np.sinc(x / np.pi)  # Ajuste para a defini√ß√£o matem√°tica usual

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="sinc(x)")
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.title("Fun√ß√£o Seno")
    plt.legend()
    plt.show()

def plot_exponential():
  #Fun√ß√£o para plotar curvas exponenciais com diferentes coeficientes
    x = np.linspace(-2, 2, 400)
    a_values = [-10, -1, 2, 10]

    plt.figure(figsize=(8, 5))
    for a in a_values:
        y = np.exp(a * x) #Equa√ß√£o da fun√ß√£o exponencial
        plt.plot(x, y, label=f"a={a}")

    plt.title("Fun√ß√£o Exponencial")
    plt.legend()
    plt.show()
#Fun√ß√£o da parabola
def plot_parabolas():
    cases = [
        (1, -3, 2),   # Discriminante positivo (duas ra√≠zes reais)
        (1, -2, 1),   # Discriminante zero (uma raiz real)
        (1, 1, 1)     # Discriminante negativo (ra√≠zes complexas)
    ]

    x = np.linspace(-5, 5, 400)
    plt.figure(figsize=(10, 6))

    for a, b, c in cases:
        y = a * x**2 + b * x + c #Equa√ß√£o da parabola
        discriminante = b**2 - 4*a*c

        if discriminante >= 0:
            r1 = (-b + np.sqrt(discriminante)) / (2 * a)
            r2 = (-b - np.sqrt(discriminante)) / (2 * a)
            plt.scatter([r1, r2], [0, 0], color='red', label=f"Ra√≠zes: {r1:.2f}, {r2:.2f}")

        plt.plot(x, y, label=f"a={a}, b={b}, c={c}")

    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.title("Par√°bolas e suas ra√≠zes")
    plt.legend()
    plt.show()
#Chamando as fun√ß√µes para plotar os gr√°ficos:
plot_gaussians()
plot_sinc()
plot_exponential()
plot_parabolas()
```

# Atividades avan√ßadas: 15 pts

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_deltoid():
    theta = np.linspace(0, 2 * np.pi, 400) #Cria 400 pontos igualmente espa√ßados
    x = 2 * np.cos(theta) + np.cos(2 * theta) #F√≥rmula para x da curva deltoide
    y = 2 * np.sin(theta) - np.sin(2 * theta)  #F√≥rmula para y da curva deltoide

    plt.figure(figsize=(6, 6)) #Define o tamanho do gr√°fico
    plt.plot(x, y, color='purple', label="Deltoid")  # Plota a curva
    plt.xlabel("x") # Nomeia o eixo x
    plt.ylabel("y") # Nomeia o eixo y
    plt.title("Curva Deltoid") #Adiciona um t√≠tulo
    plt.legend() #Adiciona legenda
    plt.grid() # adiciona grade
    plt.show() #exibe o gr√°fico

def plot_galileo_spiral():
    theta = np.linspace(0, 10 * np.pi, 400)
    r = theta**2
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(6, 6))
    plt.plot(x, y, color='purple', label="Espiral de Galileu")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Espiral de Galileu")
    plt.legend()
    plt.grid()
    plt.show()

def plot_fey_function():
    theta = np.linspace(0, 24 * np.pi, 400)
    r = np.exp(np.cos(theta)) - 2 * np.cos(4 * theta) + np.sin(theta / 12)**5
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(6, 6))
    plt.plot(x, y, color='purple', label="Fey‚Äôs function: ")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Fey‚Äôs function: ")
    plt.legend()
    plt.grid()
    plt.show()

# Chamando as fun√ß√µes para exibir os gr√°ficos
plot_deltoid()
plot_galileo_spiral()
plot_fey_function()
```
